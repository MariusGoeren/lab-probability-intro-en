## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
from math import factorial as fac

white = 10
red = 12
black = 8
all_balls = white + red + black

# Taking a white ball out.
prob_1st_white = white / all_balls
print(f"probability of taking a white ball first: {round(prob_1st_white,4)}")

# Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
prob_2nd_black_no_repl = black / (all_balls - 1)
print(f"probability of taking a black ball second: {round(prob_2nd_black_no_repl,4)}")
prob_1st_white_2nd_black_no_repl = prob_1st_white * prob_2nd_black_no_repl
print(f"probability of taking a white ball first and a black ball second (without replacing): {round(prob_1st_white_2nd_black_no_repl,4)}")

# Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
prob_2nd_black_repl = black / all_balls 
print(f"probability of taking a black ball first: {round(prob_2nd_black_repl,4)}")
prob_1st_white_2nd_black_repl = prob_1st_white * prob_2nd_black_repl
print(f"probability of taking a white ball first and a black ball second (with replacing): {round(prob_1st_white_2nd_black_repl,4)}")

# Taking 3 balls and them being all different colours.
possible_cases = fac(30) / (fac(3)*fac(27))
favorable_cases = white * red * black

prob_all_colors_one_time = favorable_cases / possible_cases
print(f"probability of taking every color once (3 picks): {round(prob_all_colors_one_time,4)}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
prob_cloudy = 0.4
prob_rainy_when_cloudy = 0.5
prob_rainy_not_cloudy = 0.2
prob_not_cloudy = 1 - prob_cloudy

prob_rain_on_cloudy_day = prob_cloudy * prob_rainy_when_cloudy
prob_rain_on_non_cloudy_day = prob_not_cloudy * prob_rainy_not_cloudy

prob_rainy_saturday = prob_rain_on_cloudy_day + prob_rain_on_non_cloudy_day

print(f"probability of rain during next saturday: {prob_rainy_saturday}")
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
1. $P(\overline{A}) = 0.7$
2. $P(A \cup B) = 0.6$
3. $P(A \cap \overline{B}) = 0.2$
4. $P(A \cup \overline{A}) = 1$  
5. $P(A \cap (B \cup \overline{B})) = 0.4$
6. $P(\overline{A} \cap \overline{B}) = 0.3$
7. $P(\overline{A \cap B}) = 0.9$

```python
# YOUR CODE HERE
print(f"1. result: {(1.0 - 0.3)}")
print(f"1.: {(1.0 - 0.3) == 0.7}\n")
print(f"2. result: {0.3 + 0.6 - 0.1}")                  # P(A) + P(B) -> -0.1 because you need the intersection of A and B only once
print(f"2.: {0.3 + 0.6 - 0.1 == 0.6}\n")
print(f"3. result: {0.3 - 0.1}")                        # only the part of A -> -0.1 because you don't need the intersection of A and B
print(f"3.: {0.3 - 0.1 == 0.2}\n") 
print(f"4. result: {0.3 + (1.0 - 0.3)}")                # A + (1.0 - A) -> is always 1
print(f"4.: {0.3 + (1.0 - 0.3) == 1.0}\n")
print(f"5. result: {0.3 * (0.6 + (1.0 - 0.6))}")        # 
print(f"5.: {0.3 * (0.6 + (1.0 - 0.6)) == 0.4}\n")
print(f"6. result: {1.0 - (0.3 + 0.6 - 0.1)}")          # everything (1.0) expcept P(A) + P (B) -> -0.1 because you need the intersection of A and B only once
print(f"6.: {1.0 - (0.3 + 0.6 - 0.1) == 0.3}\n")
print(f"7. result: {(1.0 - 0.1)}")                      # opposite of the given P(A^B) -> 1.0 - 0.1 = 0.9
print(f"7.: {(1.0 - 0.1) == 0.9}")
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE
P_covid_china = 0.01     # P(covid|china)
P_covid_italy = 0.2      # P(covid|italy)

P_china = 5 / 6     # P(china)
P_italy = 1 / 6     # P(italy)

# Question 1 : P(covid) ????????????????????            # Calculating the probability of getting covid
# = P(covid|china)P(china) + P(covid|italy)P(italy)
P_covid = P_covid_china * P_china + P_covid_italy * P_italy
print(f"probability of getting covid: {P_covid}")

# Question 2 : P(china|covid) ????????????????????            # Calculating the probability of going to china and getting covid 
# = P(covid|china)P(china) / P(covid)
P_china_covid = P_covid_china * P_china / P_covid
print(f"probability of going to china and getting covid: {P_china_covid}")
```
